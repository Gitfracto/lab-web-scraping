````python
names = soup.find_all("article", class_="Box-row d-flex")
person_name =[ e.h1.text.strip() for e in names]
person_name
````

````python
names2 = soup.find_all("p", class_="f4 text-normal mb-1")
a=-1
person_nick = []
for e in names2:
    a += 1
    person_nick.append(names2[a].text.strip())
person_nick
````

````python
dict_trending = dict(zip(person_nick, person_name))
print(dict_trending)
````

####  Challenge 2 - Display the trending Python repositories in GitHub

The steps to solve this problem is similar to the previous one except that you need to find out the repository names instead of developer names.

````python
# This is the url you will scrape in this exercise
url2 = 'https://github.com/trending/python?since=daily'
````

````python
html2=req.get(url2).content
html2[0]
````

````python
soup2=bs(html2, 'lxml')
soup2
````

````python
repo = soup2.find_all("h1", class_="h3 lh-condensed")
repo
name_repository =[ (f.a.text.strip("\n")).strip().split(" /\n\n      ") for f in repo]
name_repository
````

````python
repository = [g[1] for g in name_repository]
print(repository)
````

#### Challenge 5 - Number of Titles that have changed in the United States Code since its last release point

````python

html5=req.get(url5).content
soup5=bs(html5, 'lxml')
titles = soup5.find_all("div", class_="uscitem")
titles 
for e in titles:
    us_title =[e.div.text.strip() for e in titles]
us_title


us_title1 =[f for f in us_title if "Title" in f ]
us_title1
````

#### Challenge 7 - List all language names and number of related articles in the order they appear in wikipedia.org

````python
# This is the url you will scrape in this exercise
url7 = 'https://www.wikipedia.org/'
````

````python
html7=req.get(url7).content
soup7=bs(html7, 'lxml')
idioma = soup7.find_all("div", class_="central-featured-lang")
len(idioma)

a=-1
idim = []
for e in idioma:
    a += 1
    idim.append(idioma[a].text.strip())
idim


idim2 = [i.replace(u'\xa0', u'') for i in idim]
idim3 = [i.replace("\n"," ") for i in idim2]
idim3
````
